```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Na    = 2
Rf    = 0.03 
mu    = 0.08
sigma = 0.25
b     = 0.02
gamma = 1


X0       = 1
X1   = np.linspace(0.001,2,1000)
U_b   = np.empty((1000,1))
U_s   = np.empty((1000,1))
id_asset = 1 
for i in range(1000): 
    U_b[i] = Rf+X1[i]*(mu-Rf)-0.5*gamma*sigma**2*X1[i]**2-(X1[i]-X0)*b*1
    U_s[i] = Rf+X1[i]*(mu-Rf)-0.5*gamma*sigma**2*X1[i]**2-(X1[i]-X0)*b*-1

X1_star_b = (mu-Rf-b*1)/(gamma*sigma**2)
X1_star_s = (mu-Rf-b*-1)/(gamma*sigma**2)
U_star_b = Rf+X1_star_b*(mu-Rf)-0.5*gamma*sigma**2*X1_star_b**2-(X1_star_b-X0)*b*1
U_star_s = Rf+X1_star_s*(mu-Rf)-0.5*gamma*sigma**2*X1_star_s**2-(X1_star_s-X0)*b*-1
    
plt.plot(X1,U_b,'b-',label='buying risky')
plt.plot(X1_star_b,U_star_b,'bo')
plt.plot(X1,U_s,'r-',label='selling risky')
plt.plot(X1_star_s,U_star_s,'ro')
plt.plot([X1_star_b,X1_star_b],[-0.01,U_star_b],'k--')
plt.plot([X1_star_s,X1_star_s],[-0.01,U_star_s],'k--')
plt.xlabel('$X_1$')
plt.ylabel('Utility')
plt.legend()
plt.annotate('no-trade zone', xy=((X1_star_b+X1_star_s)/2,0.005), xytext=((X1_star_b+X1_star_s)/2,0.005), ha="center", va="bottom")
plt.show()
```
